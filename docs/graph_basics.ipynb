```python
import pandas as pd
```

```python
sales = [500, 670, 840, 690, 730]
years = [2018, 2019, 2020, 2021, 2022]

df = pd.DataFrame({'Sales': sales, 
                  'Year': years})
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'])
plt.show(sales_plot)
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'], color='green')
plt.show(sales_plot)
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'], color='green', linestyle="dashed")
plt.show(sales_plot)
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'], color='green', linestyle="dashed", marker="o")
plt.show(sales_plot)
```

```python
plt.plot(df["Year"], df['Sales'], color='green', linestyle="dashed", marker="o")
plt.title("Sales over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
```

## Bar Graphs

```python
import random
students = ['Student 1', 'Student 2', 'Student 3', 'Student 4']
random.seed(4)
averages = [random.randint(70, 100) for _ in range(4)]
bar_data = pd.DataFrame({'Students':students, 'Average':averages})
```

```python
bar_data.plot.bar(x='Students', y='Average')
```

```python
bar_data.plot.bar(x='Students', y='Average', alpha=.7)
plt.show()
```

```python
bar_data.plot.bar(x='Students', y='Average', alpha=.7, color='y')
plt.show()
```
