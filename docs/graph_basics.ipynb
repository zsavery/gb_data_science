```python
import pandas as pd
```

```python
sales = [500, 670, 840, 690, 730]
years = [2018, 2019, 2020, 2021, 2022]

df = pd.DataFrame({'Sales': sales, 
                  'Year': years})
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'])
plt.show(sales_plot)
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'], color='green')
plt.show(sales_plot)
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'], color='green', linestyle="dashed")
plt.show(sales_plot)
```

```python
sales_plot = plt.plot(df["Year"], df['Sales'], color='green', linestyle="dashed", marker="o")
plt.show(sales_plot)
```

```python
plt.plot(df["Year"], df['Sales'], color='green', linestyle="dashed", marker="o")
plt.title("Sales over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
```
