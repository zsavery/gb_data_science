```python
# sales report

import pandas as pd
```

```python
# import data from a csv
raw_data = pd.read_csv("data/electronic_store_sales.csv")
```

```python
# get information about data
raw_data.info()
```

```python
sellers = raw_data["Employee Sold By"]
```

```python
sellers = pd.Series(sellers)
```

```python
sold_by = {}
for name in sellers:
    if name in sold_by.keys():
        sold_by[name] += 1
    else:
        sold_by[name] = 1

print(sold_by)
```

```python
employee_sales_list = pd.DataFrame({"Employee":sold_by.keys(), "Product Sold":sold_by.values()})
```
