```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
students = ['Student A', 'Student B', 'Student C', 'Student D']
```

## Q1 - Number of daily restroom visits

### Bar Graph

```python
# Restroom data
restroom = [3, 4, 2, 1]
x = np.arange(len(restroom))
plt.bar(x,restroom, color='blue', width=0.4)
plt.xlabel('Students')
plt.ylabel('Restroom Visits')
plt.title('Restroom Visits by Student')
plt.xticks(x, students)
plt.yticks([0, 1, 2, 3, 4, 5])
plt.show()
```

## Q2 - How many hours per day do you read?

```python
# sleep data
reading_hrs = [22, 1.5, 2 , 1]
reading_df = pd.DataFrame({'Students': students, 'Sleep Hours': reading_hrs})
plt.boxplot(reading_df['Sleep Hours'], vert=False)
plt.xlabel('Sleep Hours')
plt.title('Sleep Hours by Student')
plt.yticks([1], ['Students'])
plt.show()
```

## Q4 - Who is the MVP in the NBA?

```python
mvp = ["Jordan", "Jordan", "Lebron", "Lebron"]
mvp_dict = {name: mvp.count(name) for name in mvp}

mvp_pie = plt.figure()
mvp_pie.add_axes([0.1, 0.1, 0.8, 0.8])
plt.pie(mvp_dict.values(), labels=mvp_dict.keys())
plt.title('MVP Count')
plt.show()  
```

```python
carriers = ["Xfinity", "Xfinity", "Xfinity", "Verizion"]
plt.bar(carriers, [carriers.count(c) for c in carriers], color='green')
plt.xlabel('Carriers')
plt.ylabel('Count')
plt.title('Carrier Distribution')
plt.xticks(rotation=45)
plt.show()
```

# Big O Notation

```python
n = np.arange(0., 10., .5) # all numbers form 0-10 at step of 0.5
c = [10 for _ in range(len(n))]
n_squared = n**2
n_cubed = n**3
n_log = n**1/2
n_pow_n = n**n

plt.plot(n, 'bo', n_squared, 'gs', n_cubed, 'r^', c, 'k-')
plt.xlim(0,20)
```
